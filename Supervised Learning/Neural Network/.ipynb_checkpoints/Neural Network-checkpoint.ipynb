Overall, Fashion MNIST is a valuable resource for experimenting with and developing image classification algorithms, providing a diverse set of examples in a compact and accessible format.

## Image Flattening

Basic dense neural networks take feature vectors and pass them through the initial (0-th or first) layer of the computational graph, which is structured similarly to single-neuron models. To input images into such a network, it's necessary to flatten the matrix into a column vector.



For each image matrix under consideration in the Fashion MNIST dataset, we can achieve this by applying the ```flatten()``` method along with the ```reshape(784, 1)``` method to ensure it becomes a column vector. It's important to note that each image matrix is originally 28 by 28, resulting in $784 = 28 \times 28$ as the number of rows in the flattened matrix column vector. The numerical values in the flattened training and testing data matrices range between 0 and 255. Due to the significant differences in potential values, this can pose challenges during the training of neural network weights and biases. A quick remedy is to scale all data to fit within the interval $(0, 1)$ by dividing each entry by the maximum possible value, which in this case is 255.

## One-Hot Encoding

The subsequent code normalizes our training and testing data, transforms our images through reshaping, and saves them using new variable names. Additionally, it performs one-hot encoding on the labels.

````python
# Scale the Fashion MNIST data for training.
scaled_train_X = train_X / 255
scaled_test_X = test_X / 255

# Flatten the training images into column vectors.
flat_train_X = []

# One-hot encode the training labels
onehot_train_y = []

for image, label in zip(scaled_train_X, train_y):
    flat_train_X.append(image.flatten().reshape(784, 1))
    temp_vec = np.zeros((10, 1))
    temp_vec[label][0] = 1.0
    onehot_train_y.append(temp_vec)

# Process the testing data in a similar manner.
flat_test_X = []
onehot_test_y = []

for image, label in zip(scaled_test_X, test_y):
    flat_test_X.append(image.flatten().reshape(784, 1))
    temp_vec = np.zeros((10, 1))
    temp_vec[label][0] = 1.0
    onehot_test_y.append(temp_vec)
````

## Building the Network Architecture

For our objectives, we aim to construct a multi-layered fully connected, or **dense**, neural network comprising $L$ layers. The network will have 784 input nodes, $L-2$ hidden layers of arbitrary size, and 10 output nodes.

The Activation Function will be the Sigmoid Function:

$$
\sigma(z) = \frac{1}{1+e^{-z}}
$$

The Cost Function will be the Mean Sqaure Error cost:

$$
C(W, b) = \frac{1}{2}\sum_{k=1}^{10}(\hat{y}^{(i)}_k - y^{(i)}_k)^2
$$

Our objective is to create a custom Python class that implements our desired structure. However, before proceeding with that, we will first sequentially write functions to gain a better understanding of the process of programming the following:

* Initialization of the weights and biases for each layer
* The forward pass
* Computation of the cost function
* Computation of the gradient
* Iteration of stochastic gradient descent

Initially, we'll establish our sigmoid activation function, its derivative, and the mean squared error for a single instance of training data. Execute the subsequent code to accomplish this.

````python
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

def d_sigmoid(z):
    return sigmoid(z) * (1.0 - sigmoid(z))

def mean_squared_error(a, y):
    return 0.5 * sum(np.sum((a[i] - y[i]) ** 2) for i in range(10))
````

Following that, we'll create a custom function to initialize the weight matrices and bias column vectors for a dense neural network. Execute the provided code to implement this.

````python
def initialize_weights(layers=[784, 60, 60, 10]):
    """
    Initialize weight matrices and bias column vectors for a dense neural network.

    Parameters:
    - layers (list): List specifying the number of nodes in each layer of the neural network.

    Returns:
    - W (list): List of weight matrices for each layer.
    - B (list): List of bias column vectors for each layer.
    """

    # Lists to contain numpy matrices connecting the layers in the neural network
    W = [[0.0]]
    B = [[0.0]]

    # Loop through each layer to initialize weight matrices and bias vectors
    for i in range(1, len(layers)):
        # "The scaling factor is a heuristic found in a research paper" - from lecture
        w_temp = np.random.randn(layers[i], layers[i-1]) * np.sqrt(2 / layers[i-1])
        b_temp = np.random.randn(layers[i], 1) * np.sqrt(2 / layers[i-1])

        W.append(w_temp)
        B.append(b_temp)

    return W, B
````

# Feedforward Phase

The Feedforward Phase, also known as the forward pass, is a crucial step in the operation of a neural network. It involves the propagation of input data through the network to generate predictions or activations at the output layer. During this phase, each layer in the network computes its weighted sum of inputs, applies an activation function to produce its output, and passes this output as input to the next layer.

The Feedforward Phase is an essential component of training and making predictions with neural networks. During training, the predicted output is compared to the actual target values, and the network's parameters (weights and biases) are adjusted in the direction that reduces the prediction error, typically through a process called backpropagation.

For each layer $\ell$ in our network, where $\ell$ ranges from $1$ to $L$, there are two distinct phases: the preactivation phase given by $$\mathbf{z}^{\ell} = W^{\ell}\mathbf{a}^{\ell-1} + \mathbf{b}^{\ell},$$ and the postactivation phase defined as $$\mathbf{a}^{\ell} = \sigma(\mathbf{z}^{\ell}).$$ In the preactivation phase, we compute a weighted linear combination of the postactivation values from the preceding layer. Subsequently, in the postactivation phase, each element of this preactivation value undergoes transformation through an activation function. It's important to note that $\mathbf{a}^0 = \mathbf{x}^{(i)}$, where $\mathbf{x}^{(i)}$ represents the current input data fed into our network.

We can validate our activation functions and matrix dimensions by executing the following code, which manually simulates the feedforward process on a neural network with the specified dimensions.

````python
# Initialize weights and biases
W, B = initialize_weights()

# Select a sample from the training set
xi = flat_train_X[0]
yi = onehot_train_y[0]
a0 = xi

# Display the shape of the input layer
print(f"Shape of the input layer (a0): {np.shape(a0)} \n")

# First hidden layer
z1 = W[1] @ a0 + B[1]
a1 = sigmoid(z1)

# Display the shapes of weights, weighted sum, and activation for the first hidden layer
print(f"Shape of weights for the first hidden layer (W[1]): {np.shape(W[1])}")
print(f"Shape of weighted sum for the first hidden layer (z1): {np.shape(z1)}")
print(f"Shape of activation for the first hidden layer (a1): {np.shape(a1)} \n")

# Second hidden layer
z2 = W[2] @ a1 + B[2]
a2 = sigmoid(z2)

# Display the shapes of weights, weighted sum, and activation for the second hidden layer
print(f"Shape of weights for the second hidden layer (W[2]): {np.shape(W[2])}")
print(f"Shape of weighted sum for the second hidden layer (z2): {np.shape(z2)}")
print(f"Shape of activation for the second hidden layer (a2): {np.shape(a2)} \n")

# Output layer
z3 = W[3] @ a2 + B[3]
a3 = sigmoid(z3)
y_hat = a3

# Display the shapes of weights, weighted sum, and activation for the output layer
print(f"Shape of weights for the output layer (W[3]): {np.shape(W[3])}")
print(f"Shape of weighted sum for the output layer (z3): {np.shape(z3)}")
print(f"Shape of activation for the output layer (a3): {np.shape(a3)} \n")

# Display prediction, target label, and mean squared error
print(f"Prediction: np.argmax(y_hat) = {np.argmax(y_hat)}")
print(f"Target Label: np.argmax(yi) = {np.argmax(yi)}")
print(f"Mean Squared Error (MSE): {mean_squared_error(y_hat, yi)}")
````

Naturally, we have the option to encapsulate these lines of code within a ```for```-loop and encapsulate it in a callable function. The provided code accomplishes exactly that.

````python
def forward_pass(W, B, xi, predict_vector=False):
    """
    Perform a forward pass through a neural network.

    Parameters:
    - W (list): List of weight matrices for each layer.
    - B (list): List of bias column vectors for each layer.
    - xi (numpy.ndarray): Input vector for the forward pass.
    - predict_vector (bool): If True, return the final prediction vector; if False, return intermediate results.

    Returns:
    - tuple or numpy.ndarray: If predict_vector is False, return a tuple (Z, A) containing weighted sums and activations for each layer.
                              If predict_vector is True, return the final prediction vector.
    """
    
    Z = [[0.0]]
    A = [xi]
    L = len(W) - 1

    for i in range(1, L + 1):
        # Calculate the weighted sum for the current layer
        z = W[i] @ A[i-1] + B[i]
        Z.append(z)
        
        # Apply the sigmoid activation function
        a = sigmoid(z)
        A.append(a)
        
    if not predict_vector:
        return Z, A
    else:
        return A[-1]

def predict(W, B, xi):
    """
    Make a prediction using the forward pass of a neural network.

    Parameters:
    - W (list): List of weight matrices for each layer.
    - B (list): List of bias column vectors for each layer.
    - xi (numpy.ndarray): Input vector for prediction.

    Returns:
    - int: Predicted class label.
    """
    
    # Perform a forward pass to get the prediction vector
    _, A = forward_pass(W, B, xi)
    return np.argmax(A[-1])

# Example usage
y_hat = forward_pass(W, B, flat_train_X[0], predict_vector=True)
print(f"Prediction: np.argmax(y_hat) = {np.argmax(y_hat)}")
print(f"Target Label: np.argmax(yi) = {np.argmax(yi)}")
print(f"Mean Squared Error (MSE): {mean_squared_error(y_hat, yi)}")
````

Next, we will create a specialized function to generate predictions for a random data point. Additionally, we will craft a mean squared error function that calculates the error across an entire set of features and labels.

````python
def random_experiment(W, B, data_features, data_labels):
    """
    Perform a random experiment by making predictions on a randomly selected data point.

    Parameters:
    - W (list): List of weight matrices for each layer.
    - B (list): List of bias column vectors for each layer.
    - data_features (list): List of input features.
    - data_labels (list): List of corresponding labels.
    """
    
    # Select a random index from the dataset
    i = np.random.randint(len(data_features))
    
    # Print the actual and predicted labels for the selected data point
    print(f"Actual label: {np.argmax(data_labels[i])}")
    print(f"Predicted label: {predict(W, B, data_features[i])}")

def MSE(W, B, X, y):
    """
    Calculate the Mean Squared Error (MSE) for a set of features and labels.

    Parameters:
    - W (list): List of weight matrices for each layer.
    - B (list): List of bias column vectors for each layer.
    - X (list): List of input features.
    - y (list): List of corresponding labels.

    Returns:
    - float: Mean Squared Error (MSE) for the dataset.
    """
    
    # Initialize variables for accumulating the total cost and the count of data points (m)
    cost = 0.0
    m = 0

    # Loop through each data point and compute the MSE
    for xi, yi in zip(X, y):
        # Perform a forward pass to get the prediction vector
        a = forward_pass(W, B, xi, predict_vector=True)
        
        # Accumulate the squared error for each data point
        cost += mean_squared_error(a, yi)
        m += 1

    # Calculate the mean error over the entire dataset
    return cost / m

# Calculate and print the Mean Squared Error for the training dataset
mse_train = MSE(W, B, flat_train_X, onehot_train_y)
print(f"Mean Squared Error on Training Data: {mse_train} \n")

# Perform a random experiment on the training dataset
random_experiment(W, B, flat_train_X, onehot_train_y)
````

## Backpropogation Phase with Stochastic Gradient Descent

We are prepared to create a custom Python ```DenseNetwork``` class. This class will handle the initialization of weights and biases for the network and implement stochastic gradient descent, as demonstrated below.

1. Initialize Parameters
    * Initialize the weights (W) and biases (B) randomly.
2. Forward Pass:
    * For a randomly selected training example ($x_i$, $y_i$):
        * Compute the predicted output ŷ using the current weights and biases.
        * Calculate the loss or error between ŷ and the actual label $y_i$.
3. Backward Pass (Backpropagation):
    * Compute the gradient of the loss with respect to the weights and biases.
    * Update the weights and biases in the opposite direction of the gradient to minimize the loss.   
   Mathematically, the update for each weight $W_{ij}$ and bias $B_i$ in layer $l$ is given below, where α is the learning rate and J is the loss function: 

$$
W^{\ell}_{ij} \leftarrow W^{\ell}_{ij} - \alpha \frac{\delta J}{\delta W^{\ell}_{ij}}
$$
$$
B^{\ell}_{i} \leftarrow B^{\ell}_{i} - \alpha \frac{\delta J}{\delta B^{\ell}_{i}}
$$

4. Repeat:
    * Repeat steps 2-3 for a specified number of iterations or until convergence.
    
The key idea is to update the model parameters using the gradient of the loss with respect to those parameters. The stochastic aspect comes from the fact that each update is performed on a single randomly chosen training example, making the process more computationally efficient compared to traditional Gradient Descent.

````python
class DenseNetwork(object):
    """
    Custom implementation of a dense neural network using stochastic gradient descent for training.

    Parameters:
    - layers (list): List specifying the number of nodes in each layer.
    """

    def __init__(self, layers=[784, 60, 60, 10]):
        """
        Initialize the DenseNetwork object.

        Parameters:
        - layers (list): List specifying the number of nodes in each layer.
        """
        
        self.layers = layers
        self.W, self.B = initialize_weights(layers=self.layers)

    def train(self, X_train, y_train, alpha=0.046, epochs=4):
        """
        Train the neural network using stochastic gradient descent.

        Parameters:
        - X_train (list): List of input features for training.
        - y_train (list): List of corresponding labels for training.
        - alpha (float): Learning rate for gradient descent.
        - epochs (int): Number of training epochs.
        """
        
        # Print the initial mean squared error
        self.errors_ = [MSE(self.W, self.B, X_train, y_train)]
        print(f"Starting Cost = {self.errors_[0]}")

        # Find your sample size
        sample_size = len(X_train)

        # Find the number of non-input layers.
        L = len(self.layers) - 1

        # For each epoch perform stochastic gradient descent.
        for k in range(epochs):
            # Loop over each (xi, yi) training pair of data.
            for xi, yi in zip(X_train, y_train):
                # Use the forward pass function defined before
                # and find the preactivation and postactivation values.
                Z, A = forward_pass(self.W, self.B, xi)

                # Store the errors in a dictionary for clear interpretation
                # of computation of these values.
                deltas = dict()

                # Compute the output error
                output_error = (A[L] - yi) * d_sigmoid(Z[L])
                deltas[L] = output_error

                # Loop from L-1 to 1. Recall the right entry of the range function
                # is non-inclusive.
                for i in range(L - 1, 0, -1):
                    # Compute the node errors at each hidden layer
                    deltas[i] = (self.W[i + 1].T @ deltas[i + 1]) * d_sigmoid(Z[i])

                # Loop over each hidden layer and the output layer to perform gradient descent.
                for i in range(1, L + 1):
                    self.W[i] -= alpha * deltas[i] @ A[i - 1].T
                    self.B[i] -= alpha * deltas[i]

            # Show the user the cost over all training examples
            self.errors_.append(MSE(self.W, self.B, X_train, y_train))
            print(f"{k + 1}-Epoch Cost = {self.errors_[-1]}")

    def predict(self, xi):
        """
        Make a prediction using the trained neural network.

        Parameters:
        - xi (numpy.ndarray): Input vector for prediction.

        Returns:
        - int: Predicted class label.
        """
        
        depth = len(self.layers)
        _, A = forward_pass(self.W, self.B, xi)
        return np.argmax(A[-1])
````

The cell below will take rougly 7-10 minutes to run through all the epoch costs.

````python
# Initialize a network with 784 input nodes, two hidden layers with 120 and 145 nodes respectively,
# and an output layer with 10 nodes.
net = DenseNetwork(layers=[784, 120, 145, 120, 10])

# Train the neural network on the training data
# Parameters:
# - flat_train_X: List of flattened training images as input features.
# - onehot_train_y: List of one-hot encoded training labels.
# - alpha: Learning rate for stochastic gradient descent.
# - epochs: Number of training epochs (iterations through the entire training dataset).
net.train(flat_train_X, onehot_train_y, alpha=0.046, epochs=4)

# Check the mean squared error after training
final_mse = MSE(net.W, net.B, flat_train_X, onehot_train_y)
print(f"Final Mean Squared Error on Training Data: {final_mse} \n")

# Make a random prediction after training
random_index = np.random.randint(0, len(flat_test_X))
prediction = net.predict(flat_test_X[random_index])
actual_label = np.argmax(onehot_test_y[random_index])

print(f"Predicted Digit: {prediction}")
print(f"Actual Digit: {actual_label}")

# Display the corresponding image
plt.figure(figsize=(8, 8))
plt.imshow(test_X[random_index], cmap="gray")
plt.show()
````
